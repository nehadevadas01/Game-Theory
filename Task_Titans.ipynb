```python
import matplotlib.pyplot as plt
import numpy as np

def plot_two_lines(a, c1, c2):
    """
    Plots two lines given parameters and calculates their intersection point.

    Parameters:
    a (float): An arbitrary parameter affecting the slope or intercept.
    c1 (float): Parameter affecting the first line.
    c2 (float): Parameter affecting the second line.
    """
    # Generate x values
    x = np.linspace(0, 100, 10000)  # 400 points between 0 and 100

    # Compute y values for each line
    y1 = (-1/2) * x + (1/2) * a + (-1/2) * c1
    y2 = (x + (1/2) * c2 - (1/2) * a) * (-2)

    # Calculate the intersection point
    # Solve (-1/2)x + (1/2)a - (1/2)c1 = (x + (1/2)c2 - (1/2)a)*(-2)
    # Rearranging, we get x_intersect = (c1 - c2 + a) / (3/2)
    x_intersect = (c1 - 2*c2 + a) / 3
    y_intersect = (c2 - 2*c1 + a) / 3

    # Create a new figure
    plt.figure(figsize=(8, 6))

    # Plot the first line
    plt.plot(x, y1, color='r', label='Line 1')

    # Plot the second line
    plt.plot(x, y2, color='b', label='Line 2')

    # Plot the intersection point
    plt.scatter(x_intersect, y_intersect, color='green', zorder=5)
    plt.annotate(f'({x_intersect:.2f}, {y_intersect:.2f})',
                 xy=(x_intersect, y_intersect), 
                 xytext=(x_intersect , y_intersect ),
                 textcoords='offset points',
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 fontsize=12)

    # Add title and labels
    plt.title('Plot of Two Lines with Intersection Point')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Add a legend
    plt.legend()

    # Add grid
    plt.grid(True)

    # Display the plot
    plt.show()

# Example usage
t1 = float(input("Enter tasks completed by player 1 t1: "))
c1 = float(input("Enter health lost for player 1 c1: "))

t2 = float(input("Enter tasks completed by player 2 t2: "))
c2 = float(input("Enter health lost for player 2 c2: "))

a = float(input("Enter parameter a: "))

t = t1 + t2
p = a-t
print("Points earned by player 1 = ", t1 * p - t1 * c1)
print("Points earned by player 2 = ", t2 * p - t2 * c2)

br_t1= (a-t2-c1)/2
br_t2= (a-t1-c2)/2
print("Best response i.e given player 2 completes t2 tasks how many tasks player 1 should do get higher profit : " , br_t1)
print ("Points earned by player 1 if above tasks are completed : ", (br_t1*p)-br_t1*c1)
print("Best response i.e given player 1 completes t1 tasks how many tasks player 2 should do get higher profit : ", br_t2)
print("Points earned by player 2 if above tasks are completed : " , (br_t2*p)-br_t2*c1)
m_br_t1= (a+c1-2*c2)/3
m_br_t2= (a+c2-2*c1)/3
print("Mutual best response (number of tasks to complete) for player 1 : " , m_br_t1)
print("Points earned by player 1 if mutually best tasks are completed : " , (m_br_t1*p)-m_br_t1*c1)
print("Mutual best response (number of tasks to complete) for player 2 : " , m_br_t2)
print("Points earned by player 2 if mutually best tasks are completed : ", (m_br_t2*p)-m_br_t2*c2)

plot_two_lines(a, c1, c2)
```
